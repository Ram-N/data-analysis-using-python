### We can use many of the `STRING` operations on Columns

Let's look at this very briefly

```python
cond = countries['Name'].str.contains('bia')
countries[cond]
```

```python
#Whenever you have a condition, create a "MASK". Then apply that MASK to the data frame...
condition = countries['Name'].str.contains("bia")
small_df = countries[condition]
print(small_df['Name'])
```

### How did this work?

- First, we created a condition. In this case the `condition` was all the rows in the column called 'Name' that contained the substring "bia."
- If you print out 'condition' you will see that it is a long series of 0 and 1. (Called a MASK)
- Next we use that mask to filter out all the unwanted rows and keep only the wanted rows.
- `countries[condition]` is a very small dataframe. All the country Names that didn't satisfy this condition were removed.
- In fact, this dataframe has only 6 rows

- The last thing we need to do is to print out the Names.
- `small_df` has many columns. But we just print out the one we are interested in
