```python
#This is a nice way to find out what your current_working_directory(cwd) is...
import os
os.getcwd()
```

```python
#this is a very Pythonic way to find out all the files in the directory
os.listdir()
```

```python
#now we can store this in a Python list and do anything with it...
all_my_files = os.listdir()

for m in all_my_files:
    if m.startswith("D"):
        print(m)
```

```python
#to combine paths, just use join separated by commas
os.path.join(os.getcwd(), "data")
os.path.join(os.getcwd(), "data", "Crude Oil Data")

#Notice that windows in adding two(!) backslashes. This was causing us problems yesterday.
```
